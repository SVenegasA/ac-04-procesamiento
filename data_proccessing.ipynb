```python
import csv
import json
import re

lista_airport_city = [] # Aquí se almacenarán las líneas antes de imprimirlas en pantalla al final

try:
    with open('nodes.csv', 'r', encoding="utf-8") as archivo_csv:
        lector_csv = csv.reader(archivo_csv)
        # Aquí abrimos el archivo de salida "airport_city.txt"
        # Si quieres que se llame "nodes2.txt", simplemente cambia el nombre aquí:
        with open("nodes2.txt", "w", encoding="utf-8") as archivo_salida:
            for fila in lector_csv:
                try:
                    # Saltar filas que no sean de datos (ej. encabezados o marcadores)
                    if fila[0] == "[Airport]":
                        pass
                    elif fila[0] == "[Country]":
                        break
                    else:
                        continue

                    if len(fila) > 2:
                        id_airport = fila[1].strip()
                        diccionario_str_original = fila[2].strip()

                        contenido = diccionario_str_original.strip('{}')

                        # Buscamos 'city' o 'code' directamente con expresiones regulares
                        match_city = re.search(r'city:\s*([^,}]+)', diccionario_str_original)
                        city_name = None
                        if match_city:
                            city_name = match_city.group(1).strip().strip("'\"")

                        match_code = re.search(r'code:\s*([^,}]+)', diccionario_str_original)
                        code_value = None
                        if match_code:
                            code_value = match_code.group(1).strip().strip("'\"")

                        if city_name:
                            line_to_save = f"{id_airport} {city_name}\n"
                            lista_airport_city.append(line_to_save)
                            # ¡Esta es la línea que guarda en el archivo!
                            archivo_salida.write(line_to_save)
                        elif code_value:
                            line_to_save = f"{id_airport} {code_value}\n"
                            lista_airport_city.append(line_to_save)
                            # ¡Esta es la línea que guarda en el archivo!
                            archivo_salida.write(line_to_save)
                        else:
                            print(f"Advertencia: No se encontró 'city' ni 'code' en el diccionario de la fila: {fila}")
                    else:
                        print(f"Advertencia: Fila con menos de 3 columnas, se omite: {fila}")

                except IndexError:
                    print(f"Error de índice: La fila {fila} no tiene suficientes columnas esperadas.")
                except Exception as e:
                    print(f"Ocurrió un error inesperado al procesar la fila {fila}: {e}")

    print("\nProceso completado. Contenido de las líneas guardadas en 'lista_airport_city':")
    for item in lista_airport_city:
        print(item.strip())


except FileNotFoundError:
    print("Error: El archivo 'nodes.csv' no se encontró. Asegúrate de que está en la misma carpeta o proporciona la ruta completa.")
except Exception as e:
    print(f"Ocurrió un error general al abrir o procesar el archivo CSV: {e}")
```

```python
import csv
import json
import re

lista_airport_city = []

try:
    with open('nodes.csv', 'r', encoding="utf-8") as archivo_csv:
        lector_csv = csv.reader(archivo_csv)
        with open("nodes2.txt", "w", encoding="utf-8") as archivo_salida:
            for i, fila in enumerate(lector_csv): # Añadimos 'i' para el número de fila
                if i < 5: # Solo para las primeras 5 filas para no saturar la consola
                    print(f"\n--- Procesando Fila {i+1} ---")
                    print(f"Fila completa: {fila}")

                try:
                    if fila[0] == "[Airport]":
                        pass
                    elif fila[0] == "[Country]":
                        print("Detectado '[Country]', terminando lectura de CSV.")
                        break
                    else:
                        if i < 5:
                            print(f"Saltando fila (no '[Airport]' ni '[Country]'): {fila[0]}")
                        continue

                    if len(fila) > 2:
                        id_airport = fila[1].strip()
                        diccionario_str_original = fila[2].strip()

                        if i < 5:
                            print(f"ID Aeropuerto: '{id_airport}'")
                            print(f"String Diccionario Original: '{diccionario_str_original}'")

                        # Intentar extraer city
                        match_city = re.search(r'city:\s*([^,}]+)', diccionario_str_original)
                        city_name = None
                        if match_city:
                            city_name = match_city.group(1).strip().strip("'\"")
                            if i < 5:
                                print(f"City encontrada: '{city_name}'")

                        # Intentar extraer code (si no se encontró city)
                        code_value = None
                        if not city_name: # Solo buscar code si no se encontró city
                            match_code = re.search(r'code:\s*([^,}]+)', diccionario_str_original)
                            if match_code:
                                code_value = match_code.group(1).strip().strip("'\"")
                                if i < 5:
                                    print(f"Code encontrado: '{code_value}'")

                        if city_name:
                            line_to_save = f"{id_airport} {city_name}\n"
                            lista_airport_city.append(line_to_save)
                            archivo_salida.write(line_to_save)
                            if i < 5:
                                print(f"Línea guardada: '{line_to_save.strip()}'")
                        elif code_value:
                            line_to_save = f"{id_airport} {code_value}\n"
                            lista_airport_city.append(line_to_save)
                            archivo_salida.write(line_to_save)
                            if i < 5:
                                print(f"Línea guardada (usando code): '{line_to_save.strip()}'")
                        else:
                            if i < 20: # Solo advertir para las primeras 20 para evitar spam
                                print(f"Advertencia: No se encontró 'city' ni 'code' en la fila: {fila}")
                    else:
                        if i < 20:
                            print(f"Advertencia: Fila con menos de 3 columnas, se omite: {fila}")

                except IndexError:
                    print(f"Error de índice (fila {i+1}): La fila {fila} no tiene suficientes columnas esperadas.")
                except Exception as e:
                    print(f"Ocurrió un error inesperado (fila {i+1}): {e}. Fila: {fila}")
                    pass

    print("\n--- Proceso finalizado ---")
    print("Contenido final de lista_airport_city (primeros 10 elementos):")
    for item in lista_airport_city[:10]: # Imprimir solo los primeros 10 para no saturar
        print(item.strip())
    print(f"Total de líneas procesadas y añadidas a lista_airport_city: {len(lista_airport_city)}")


except FileNotFoundError:
    print("Error crítico: El archivo 'nodes.csv' no se encontró. Asegúrate de que está en la misma carpeta o proporciona la ruta completa.")
except Exception as e:
    print(f"Ocurrió un error general al abrir o procesar el archivo CSV: {e}")
```
